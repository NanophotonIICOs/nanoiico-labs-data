```python
import numpy as np
import matplotlib.pyplot as plt
import h5py as h5
import os

from mpl_toolkits.axes_grid1 import make_axes_locatable,ImageGrid
from mpl_toolkits.axes_grid1 import AxesGrid
from matplotlib.gridspec import GridSpec
```

```python
exppath = '/media/labfiles/lab-exps/spectro-lab/nano/grating_InP/exp_22-01/exp_13_01_afm/'
for i,files in enumerate(os.listdir(exppath)):
    print(i,files)
```

```python
exps = h5.File(exppath+'rejillatest3.h5') 
dset = list(exps.keys())
exps1 = h5.File(exppath+'testrejilla3_micro.h5') 
dset1 = list(exps1.keys())
exps2 = h5.File('/media/labfiles/lab-exps/spectro-lab/nano/grating_InP/exp_22-01/exp_14_01_afm/'+'testrejilla4_micro.h5') 
dset2 = list(exps1.keys())
```

Grating InP scan 90 micrometers using Newmark linear stages

```python
fig = plt.figure(figsize=(13, 7))
gs = GridSpec(1, 2, figure=fig,wspace=0.5,width_ratios=[2,1])

ax1 = fig.add_subplot(gs[0, 0])
ax2 =fig.add_subplot(gs[0, 1])

npix1=0
npix2=2

ax1.set_xlabel(" $x$ $(\mu m)$")
ax1.set_xticks(np.arange(0,45,4))
ax1.set_xticklabels(np.arange(0,90,8))
ax1.set_ylabel(" $y$ $(\mu m)$")
ax1.set_yticks(np.arange(0,4,2))
ax1.set_yticklabels(np.arange(0,8,4))
ax1.imshow(exps1[dset1[0]][:].T,cmap='Greys',interpolation='gaussian')
ax1.plot([0,45],[npix1,npix1],'r')
ax1.plot([0,45],[npix2,npix2],'b')

ax2.set_xlabel(" $x$ $(\mu m)$")
ax2.set_xticks(np.arange(0,45,4))
ax2.set_xticklabels(np.arange(0,90,8))
ax2.set_ylabel(" $distance ~$ $(n m)$")

ax2.plot(exps1[dset1[0]][:,npix1],'r')
ax2.plot(exps1[dset1[0]][:,npix2],'b')

plt.show()
```
